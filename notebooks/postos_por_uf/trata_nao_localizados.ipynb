```python
import os
import pandas as pd
from pathlib import Path

from dotenv import load_dotenv
load_dotenv()  # take environment variables from .env.
```

```python
# base de não localizados
file_path = os.path.join(f'nao_localizados.csv')
df_nao_localizados = pd.read_csv(file_path)
df_nao_localizados['ESTABELECIMENTO'] = df_nao_localizados['ESTABELECIMENTO'].str.strip().str.upper()
df_nao_localizados.head()
```

```python
def get_df_cnes():
  # base cnes, para pegar o código CNES da UBS
  path_folder_cnes = Path(os.getcwd()).resolve().parents[0]
  print(path_folder_cnes.parent.absolute())
  file_path = os.path.join(
    path_folder_cnes,
    'importa_csv_saude',
    'cadastro_estabelecimentos_cnes.csv')
  df_cnes = pd.read_csv(file_path, sep=";")
  df_cnes['NOME'] = df_cnes['NOME'].str.strip().str.upper()
  df_cnes = df_cnes[['CNES', 'NOME']]
  df_cnes['CNES'] = df_cnes['CNES'].fillna(0).astype(int)
  
  return df_cnes
```

```python
# liga cnes com não localizados, para incluir o código cnes
df = df_nao_localizados.merge(df_cnes, how='left', left_on='ESTABELECIMENTO', right_on='NOME')
df['CO_CNES'] = df['CNES'].fillna(0).astype(int)
del df['NOME']
df
```

```python
# liga a base de não localizados com a cnes
df = df.merge(df_cnes, how='left', on='CNES')
df['CO_CNES'] = df['CO_CNES'].fillna(0).astype(int)
df
```

```python
# base cnes
path_folder_cnes = Path(os.getcwd()).resolve().parents[0]
print(path_folder_cnes.parent.absolute())
file_path = os.path.join(
  path_folder_cnes,
  'cnes_BASE_DE_DADOS_CNES_202208',
  'cnes_estabelecimentos_202208.csv')
df_cnes = pd.read_csv(file_path)
df_cnes['CO_CEP'] = df_cnes['CO_CEP'].astype(str)
df_cnes['NO_FANTASIA'] = df_cnes['NO_FANTASIA'].str.strip().str.upper()
df_cnes['CO_CNES'] = df_cnes['CO_CNES'].fillna(0).astype(int)
print(df_cnes.info())
```

```python
# base geonames
path_folder_cnes = Path(os.getcwd()).resolve().parents[0]
print(path_folder_cnes.parent.absolute())
file_path = os.path.join(
  path_folder_cnes,
  'geonames_BR',
  'BR.txt')

columns = [
  'country_code',
  'postal_code',
  'city',
  'state',
  'nao_usado_1',
  'nao_usado_2',
  'nao_usado_3',
  'nao_usado_4',
  'nao_usado_5',
  'latitude',
  'longitude',
  'nao_usado_6'
]  
df_geonames = pd.read_csv(file_path, sep='\t', names=columns)

df_geonames = df_geonames[[
  'postal_code', 'city', 'state', 'latitude', 'longitude'
]]

df_geonames['postal_code'] = df_geonames['postal_code'].str.strip()
df_geonames['postal_code'] = df_geonames['postal_code'].str.replace('-', '', regex=False)
df_geonames.info()
```

```python
# liga a base de não localizados com a cnes
df = df.merge(df_cnes, how='left', left_on='ESTABELECIMENTO', right_on='NO_FANTASIA')
df
```
