```python
import pandas as pd
```

```python
# faz a análise dos estabelecimentos cedes, a partir do arquivo CSV
# tbEstabelecimento202208.csv (204MB), contido dentro do arquivo ZIP
# BASE_DE_DADOS_CNES_202208.ZIP (502MB), disponível para download em :
# http://cnes.datasus.gov.br/pages/downloads/arquivosBaseDados.jsp

file_path = 'tbEstabelecimento202208.csv'
df = pd.read_csv(file_path, sep=';', encoding='latin-1', low_memory=False)
print(df.columns)
df.head()

colunas = [
  'CO_CNES',
  'NO_RAZAO_SOCIAL',
  'NO_FANTASIA',
  'NO_LOGRADOURO',
  'NU_ENDERECO',
  'NO_COMPLEMENTO',
  'NO_BAIRRO',
  'CO_CEP', 
  'NU_LATITUDE',
  'NU_LONGITUDE',
]


# filtra as colunas necessárias
df = df[colunas]
df.head()
```

```python
file_path = 'cenes_estabelecimentos_202208.csv'
df.to_csv(file_path, index=False)
print('Arquivo salvo em :', file_path)
```
